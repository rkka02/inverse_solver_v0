<h3> object_center gives frequency coordinate of illumination beam in fourier space </h3>

\begin{gather*}


I(\vec{r}) = |R+U|^2 \\
\hat{I(\vec{\nu})} = \mathcal{F} [I(\vec{r})]\\
\mathcal{P} [\hat{I(\vec{\nu})}] = center[cut[\hat{I(\vec{\nu})}]]= center [\mathcal{F} [R^* U]] = \mathcal{F} [U] \\
\Downarrow \\
U(\vec{r}) = \mathcal{F}^{-1} [\mathcal{F} [U]] 


\end{gather*}

<h3> Get object field U : F[(DC term) + UR* + U*R] -> F[U] </h3>

```python
background_object_field, sample_object_field = optics.Holography_Off_Axis.get_object_field(background_hologram=background, 
                                                                                           sample_hologram=sample)
```

```python
# Option 1: Direct Subtraction
scattered_field = sample_object_field - background_object_field
```

```python
v0 = 1 / lam
n_medium = 1.33
v_nm = v0 * n_medium
k_nm = 2 * np.pi * v_nm
s_nm = v_nm // dv_ol
s_nm
```

```python
import utils

s0x = []
s0y = []
s0z = []

for i in range(Z):
    sample_freq = utils.get_maxindex(np.fft.fft2(sample_object_field[i]))
    center = np.array([N//2, N//2])
    incident_freq = sample_freq-center
    s0x.append(incident_freq[1])
    s0y.append(incident_freq[0])
    s0z.append(round(np.sqrt(s_nm**2 - incident_freq[1]**2 - incident_freq[0]**2)))
```

```python
idx = 20
temp = scattered_field[idx]
temp = np.fft.fft2(temp)
temp = np.roll(temp, shift=-s0y[idx], axis=0)
temp = np.roll(temp, shift=-s0x[idx], axis=1)
fig, axes = plt.subplots(1,2)
axes[0].imshow(np.log(1+np.abs(temp)))
axes[0].set_title('Fourier space')
axes[1].imshow(np.abs(np.fft.ifft2(temp)))
axes[1].set_title('Real space')
```

```python
s_x = np.arange(-N//2, N//2)
s_y = np.arange(-N//2, N//2)
S_x, S_y = np.meshgrid(s_x, s_y)

S_z = s_nm**2 - S_x**2 - S_y**2
S_z[S_z<0] = 0
S_z = np.sqrt(S_z)
```

```python
born = np.zeros((N, N, 300), dtype=np.complex128)
count = np.zeros((N, N, 300))
for i in tqdm(range(Z)):
    
    # Shifted coordinates
    shifted_S_x = np.roll(S_x, shift=-s0x[i], axis=1)
    shifted_S_y = np.roll(S_y, shift=-s0y[i], axis=0)
    shifted_S_z = np.roll(S_z, shift=-s0x[i], axis=1)
    shifted_S_z = np.roll(shifted_S_z, shift=-s0y[i], axis=0)
    
    # Scattered field
    Us = scattered_field[i].copy()
    fourier_Us = np.fft.fft2(Us)
    fourier_Us = np.roll(fourier_Us, shift=-s0y[i], axis=0)
    fourier_Us = np.roll(fourier_Us, shift=-s0x[i], axis=1)
    
    # Coefficients
    coeff = 2 * np.pi * dv_ol * shifted_S_z / 2 / np.pi / 1j
    
    # field
    F_born = coeff * fourier_Us

    yx_idx = np.where((S_z>0) &
                (shifted_S_x > -N//2) & (shifted_S_x < N//2) &
                (shifted_S_y > -N//2) & (shifted_S_y < N//2)
                )
    z_idx = np.round(shifted_S_z[yx_idx]-shifted_S_z.max()).astype(int)
    yxz_idx = (yx_idx[0], yx_idx[1], z_idx)
    count[yxz_idx] += np.ones((N, N, 300))[yxz_idx]
    
    born[yxz_idx] += F_born[yx_idx]  
    
born[count==0] = 0
born[count!=0] = born[count!=0]/count[count!=0]
```

```python
potential = np.fft.ifftn(np.fft.fftshift(born))
viewer.add_image(np.abs(potential))
```
